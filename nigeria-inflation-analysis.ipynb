```python
# Import libraries - these are our tools for the analysis
import pandas as pd           # for loading and working with data
import matplotlib.pyplot as plt  # for creating charts and graphs
```

```python
# Load the dataset into a dataframe, skipping the firdt 4 rows
df = pd.read_csv("API_NGA_DS2_en_csv_v2_279569.csv", skiprows=4)

# Preview the first 5 rows
df.head()
```

```python
# Check the shape of dataset - how many rows and columns 
df.shape
```

```python
# check all column names 
df.columns.tolist()
```

```python
# filter for inflation data only
inflation_df = df[df["Indicator Name"] == "Inflation,consumer prices (annual %)"]

# Preview it 
inflation_df.head()
```

```python
# see all unique indicator names
df["Indicator Name"].unique()
```

```python
# Search for any indicator containing the word "price" or "inflation" or "CPI"
df[df["Indicator Name"].str.contains("price|inflation|CPI", case=False)]["Indicator Name"].unique()
```

```python
# Filter for inflation data only
inflation_df = df[df["Indicator Name"] == "Inflation, consumer prices (annual %)"]

  #preview it 
inflation_df.head()                                                                      
```

```python
# Select only the year columns (1960 to 2025)
year_columns = [str(year) for year in range(1960, 2026)]

# Flip the data so years become rows
inflation_clean = inflation_df[year_columns].T.reset_index()

# Rename the columns
inflation_clean.columns = ["Year", "Inflation Rate"]

# Preview it
inflation_clean.head(10)
```

```python
# Convert Year and Inflation Rate to the correct data types
inflation_clean["Year"] = inflation_clean["Year"].astype(int)
inflation_clean["Inflation Rate"] = inflation_clean["Inflation Rate"].astype(float)

# Check for any missing values
print("Missing values:", inflation_clean["Inflation Rate"].isnull().sum())

# Confirm the data types look correct
print(inflation_clean.dtypes)
```

```python
# Remove rows with missing inflation data
inflation_clean = inflation_clean.dropna(subset=["Inflation Rate"])

# Confirm no more missing values
print("Missing values after cleaning:", inflation_clean["Inflation Rate"].isnull().sum())
print("Years of data we have:", len(inflation_clean))
```

```python
# plot nigeria"s inflation rate over time
plt.figure(figsize=(14, 6))
plt.plot(inflation_clean["Year"], inflation_clean["Inflation Rate"], color="blue", linewidth=2)

# Add title and labels
plt.title("Nigeria Inflation Rate (1960 - 2024)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Inflation Rate (%)", fontsize=12)

# Add a horizontal line at 0 for reference
plt.axhline(y=0, color="red", linestyle="--", linewidth=1)

plt.tight_layout()
plt.show()
```

```python
# Plot Nigeria's inflation rate over time
plt.figure(figsize=(14, 6))
plt.plot(inflation_clean["Year"], inflation_clean["Inflation Rate"], color="blue", linewidth=2)

# Highlight key historical periods
plt.axvspan(1993, 1995, color="red", alpha=0.2, label="1993-95 Crisis")
plt.axvspan(2015, 2017, color="orange", alpha=0.2, label="2016 Recession")
plt.axvspan(2020, 2024, color="yellow", alpha=0.2, label="Post-COVID Surge")

# Add titles and labels
plt.title("Nigeria Inflation Rate (1960 - 2024)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Inflation Rate (%)", fontsize=12)
plt.axhline(y=0, color="red", linestyle="--", linewidth=1)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Summary statistics for Nigeria's inflation rate
print("Highest Inflation Year:")
print(inflation_clean.loc[inflation_clean["Inflation Rate"].idxmax()])

print("\nLowest Inflation Year:")
print(inflation_clean.loc[inflation_clean["Inflation Rate"].idxmin()])

print("\nAverage Inflation Rate (1960-2024):", round(inflation_clean["Inflation Rate"].mean(), 2), "%")
print("Median Inflation Rate:", round(inflation_clean["Inflation Rate"].median(), 2), "%")
```
